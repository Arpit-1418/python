2.  Explain the difference between mutable and immutable data types with examples?
    ->data types are categorized as mutable or immutable based on whether their values can be changed after creation. Mutable data types are those that allow modifications, meaning you can add, remove, or change elements without creating a new object. Examples of mutable types include lists, dictionaries, and sets. For instance, if you have a list numbers = [1, 2, 3] and you append 4 to it, the original list itself is updated to [1, 2, 3, 4]. On the other hand, immutable data types cannot be changed once they are created. Any operation that seems to modify them actually creates a new object. Common immutable types include strings, tuples, integers, and floats. For example, if you have a string name = "Arpit" and concatenate " Kishan" to it, a new string "Arpit Kishan" is created while the original "Arpit" remains unchanged. Understanding the difference between mutable and immutable types is important because it affects how variables behave in memory and how they interact with functions, which can prevent unintended side effects in programs.

4.   Describe how dictionaries store data?
    ->dictionaries store data as key-value pairs, where each key is unique and maps to a specific value. This structure allows for fast and efficient access to data using the key rather than a numeric index, unlike lists or tuples. Internally, dictionaries use a hash table to organize the keys, which enables quick lookup, insertion, and deletion operations. Keys must be immutable types such as strings, numbers, or tuples, while values can be of any data type, including lists, tuples, or even other dictionaries. For example, a dictionary like student = {"name": "Arpit", "age": 20, "grade": "A"} stores "name", "age", and "grade" as keys, each pointing to their respective values. This structure makes dictionaries extremely useful for representing real-world data that can be identified by unique labels.

5.  Why might you use a set instead of a list in Python?
    ->In Python, you might choose to use a set instead of a list when you need a collection of unique elements and do not care about the order of items. Sets automatically eliminate duplicates, which can save time and prevent errors when working with data where uniqueness is important, such as IDs or tags. Additionally, sets are optimized for fast membership testing, so checking whether an element exists in a set is much quicker than searching through a list, especially for large collections. Sets also provide powerful operations like union, intersection, and difference, which make them ideal for tasks involving comparisons between collections. In contrast, lists maintain order and allow duplicates but are slower for membership checks and do not offer built-in support for these mathematical set operations.

14.  How does the “in” keyword work differently for lists and dictionaries?
    -> In Python, the in keyword is used to check whether an element exists in a collection, but it behaves differently for lists and dictionaries. When used with a list, it checks whether a specific value is present among the list’s elements. For example, in the list fruits = ["apple", "banana", "cherry"], the expression "apple" in fruits returns True because "apple" is one of the list items. However, when used with a dictionary, the in keyword checks only for the presence of a key, not a value. For instance, in the dictionary student = {"name": "Arpit", "age": 20}, the expression "name" in student returns True, while "Arpit" in student returns False since "Arpit" is a value, not a key. Thus, in checks for values in lists but for keys in dictionaries.

17.  Describe the time complexity of accessing elements in a dictionary
    ->In Python, accessing elements in a dictionary has an average time complexity of O(1), which means it takes constant time regardless of the dictionary’s size. This efficiency comes from the fact that dictionaries are implemented using a hash table. When you access a value using its key, Python computes the key’s hash value and directly locates the corresponding entry in memory, avoiding the need to search through all items.
